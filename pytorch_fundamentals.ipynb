# Introduction to Tensors

## Creating a Tensor:

```python
# scalar
scalar = torch.tensor(8)
scalar
```

```python
# find the dimension of the tensor
scalar.ndim
```

```python
# get the item from the tensor (converted to int)
scalar.item()
```

```python
# 1D Tensor (Vector)
vec = torch.tensor([1, 2, 3])
print(vec)

# 2D Tensor (Matrix)
m = torch.tensor([[1, 2], [3, 4]])
print(m)

# 3D Tensor (Tensor)
t = torch.tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(t)
```

## `.ndim` & `.shape` Property:

### 1. `.ndim` Property:

It just tells **how many dimentions (axes)** a tensor has.

### 2. `.shape` Property:

It tells the **exact size** of each dimension.

```python
print(vec.ndim) # couting no. of pairs of square brackets
print(m.ndim)
print(t.ndim)
```

```python
print(vec.shape)
print(m.shape)
print(t.shape)
```
